````python
import torch
from PIL import Image
import matplotlib.pyplot as plt
from torchvision import transforms

# 모델 경로 설정
generator_model_path = '/content/drive/MyDrive/sketch2face/latest_net_G_sap.pth'
deform_model_path = '/content/drive/MyDrive/sketch2face/deform.pth'
input_image_path = '/content/drive/MyDrive/sketch2face/image.jpg'

# 장치 설정 (CUDA가 사용 가능한 경우 GPU 사용)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# 전체 모델을 로드
generator_model = torch.load(generator_model_path, map_location=device)
deform_model = torch.load(deform_model_path, map_location=device)

# 모델을 평가 모드로 설정
generator_model.eval()
deform_model.eval()

# 이미지 전처리 (모델에 맞게 크기와 변환)
input_image = Image.open(input_image_path).convert("RGB")
transform = transforms.Compose([
    transforms.Resize((256, 256)),  # 모델 입력 크기에 맞게 조정
    transforms.ToTensor(),  # 텐서로 변환
    transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])  # 정규화
])

# 이미지 변환
input_tensor = transform(input_image).unsqueeze(0).to(device)  # 배치 차원 추가

# 모델을 이용해 예측 (generator_model 또는 deform_model에 따라 선택)
with torch.no_grad():
    output = generator_model(input_tensor)  # 예시: generator_model에 입력

# 출력 결과를 이미지로 변환
output_image = output.squeeze(0).cpu().numpy()  # 배치 차원 제거하고 numpy 배열로 변환
output_image = (output_image.transpose(1, 2, 0) + 1) / 2  # 정규화 복원

# 결과 이미지 출력
plt.imshow(output_image)
plt.axis('off')
plt.show()
````

````python
# 입력 이미지 로드
input_image = Image.open(input_image_path).convert('RGB')
input_image = input_image.resize((256, 256))  # 모델 입력 크기에 맞게 크기 조정

# 이미지 전처리 (Tensor로 변환)
transform = transforms.ToTensor()  # torchvision.transforms.ToTensor()로 수정
input_tensor = transform(input_image).unsqueeze(0).to(device)  # 배치 차원을 추가하고 device에 맞게 이동

# deform_model을 이용해 스케치 변형
with torch.no_grad():
    deformed_image = deform_model(input_tensor)  # deform_model이 입력 이미지를 변형한다고 가정

# 얼굴 생성 (Generator 모델을 사용)
with torch.no_grad():
    model_output = generator_model(deformed_image)  # Generator로 얼굴 생성
````

````python
# 생성된 이미지 저장
generated_image = model_output.squeeze(0).cpu().detach().numpy()
generated_image = (generated_image * 255).astype(np.uint8)  # Tensor를 이미지로 변환

# 결과 이미지를 저장
output_image_path = os.path.join(output_dir, 'generated_image.jpg')
Image.fromarray(generated_image).save(output_image_path)

# 결과 이미지 출력
output_image = Image.open(output_image_path)
plt.imshow(output_image)
plt.axis('off')  # 축 제거
plt.show()

print("Processing complete. Generated image saved at:", output_image_path)
````

````python
import torch
import os
import numpy as np
import matplotlib.pyplot as plt
from torchvision import transforms
from PIL import Image
from google.colab import drive

# Google Drive 마운트
drive.mount("/content/drive")

# 모델 경로 설정
generator_model_path = '/content/drive/MyDrive/sketch2face/latest_net_G.pth'

# GPU 사용 여부 확인
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# 모델 정의 (단일 생성기 모델을 로드하기 위한 예시)
class GeneratorModel(torch.nn.Module):
    def __init__(self):
        super(GeneratorModel, self).__init__()
        self.layer = torch.nn.Sequential(
            torch.nn.Conv2d(1, 64, kernel_size=3, stride=1, padding=1),  # 채널 수를 1로 변경 (흑백 스케치)
            torch.nn.ReLU(),
            torch.nn.Conv2d(64, 3, kernel_size=3, stride=1, padding=1)
        )

    def forward(self, x):
        return self.layer(x)

# 모델 객체 생성
generator_model = GeneratorModel().to(device)

# 모델에 state_dict 로드 (strict=False를 사용하여 불일치하는 키를 무시)
generator_state_dict = torch.load(generator_model_path, map_location=device)
generator_model.load_state_dict(generator_state_dict, strict=False)

# 모델을 평가 모드로 설정
generator_model.eval()

# 이미지 불러오기 및 전처리
input_image_path = '/content/drive/MyDrive/sketch2face/image.jpg'  # 입력 스케치 이미지 경로
image = Image.open(input_image_path).convert('L')  # 이미지를 'L'로 변환하여 흑백 이미지로 변환

# 전처리 파이프라인 (Normalize 추가)
preprocess = transforms.Compose([
    transforms.Resize((256, 256)),
    transforms.ToTensor(),  # 이미지를 Tensor로 변환
    transforms.Normalize(mean=[0.5], std=[0.5])  # 정규화 범위 [-1, 1]로 조정
])

input_tensor = preprocess(image).unsqueeze(0).to(device)

# 생성된 이미지 생성
with torch.no_grad():
    generated_image = generator_model(input_tensor)

# 생성된 이미지를 3D 이미지로 변환 (RGB 이미지로 만들어줌)
generated_image = generated_image.squeeze().cpu().numpy()  # (C, H, W)
generated_image = np.clip(generated_image, 0, 1)  # Normalize the values between 0 and 1

# 2D 이미지를 3채널(RGB) 이미지로 변환 (C, H, W -> H, W, C)
generated_image_rgb = np.transpose(generated_image, (1, 2, 0))

# 생성된 이미지가 [0, 1] 범위에 있으므로 [0, 255]로 변환
generated_image_rgb = (generated_image_rgb * 255).astype(np.uint8)

# 생성된 이미지 확인
plt.imshow(generated_image_rgb)
plt.show()

# 이미지 저장
output_dir = '/content/drive/MyDrive/sketch2face/output'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

output_path = os.path.join(output_dir, 'generated_image.jpg')
plt.imsave(output_path, generated_image_rgb)

# 생성된 이미지 경로 출력
print(f"Generated image saved to: {output_path}")
````

from options.test_options import TestOptions
